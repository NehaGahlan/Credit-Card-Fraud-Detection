# 2. Feature Selection


The phase that will be performed in this report is **Features Selection**.

It is the process in which attributes are selected that can predict a variable more accurately, thus eliminating the irrelevant ones that diminish the value of the accuracy and the quality of the model itself.

The first step that will be performed is the Correlation Matrix between the continuous variables, so without the target variable (Class).

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import NullFormatter
import numpy as np
from sklearn.decomposition import PCA
from numpy.random import rand
from scipy import stats
```

## 2.1 Prepare Data

```python
dataset = pd.read_csv('../1. Exploration Data Analysis and Data Visualization/creditcard.csv', sep=",")
```

```python
datasetForCorrelation = dataset.drop(['Class'], axis=1)
```

## 2.2 Correlation Matrix with continuous variables

```python
# Correlation matrix

corrmat = datasetForCorrelation.corr()


fig = plt.figure(figsize = (20, 25))

sns.heatmap(corrmat, vmax=1.0, center=0, fmt='.2f', cmap='PuRd_r',
            square=True, linewidths=.5, annot=True, cbar_kws={"shrink": .70})
plt.show()
```

The only values that could be linked to possible correlations between features are: 

* -0.53 between V2 and Amount
* 0.40 between V7 and Amount




## 2.3 Correlation Matrix by distinguishing the two types of Class

```python
#correlation matrix 
f, (ax1, ax2) = plt.subplots(1,2,figsize =( 15, 7 ))

sns.heatmap(dataset.query('Class==1').drop(['Class','Time'],1).corr(), vmax = .8, square=True, ax = ax1, cmap = 'YlGnBu')
ax1.set_title('Fraud')

sns.heatmap(dataset.query('Class==0').drop(['Class','Time'],1).corr(), vmax = .8, square=True, ax = ax2, cmap = 'YlGnBu');
ax2.set_title('Normal')

plt.show()
```
